```python
import statbotics
import json
import time
import matplotlib.pyplot as plt
from operator import itemgetter

```

```python
sb = statbotics.Statbotics()

with open('2023champs.json') as f:
    data = json.load(f)

divisions = {}
```

```python
for division, teams in data.items():
    print("-----", division)
    epas = []
    for team in teams:
        print(team)
        while True:
            epa = sb.get_team_year(team, year=2023, fields=["epa_end"])["epa_end"]
            if epa is not None:
                break
        epas.append((epa, team))
    epas = list(zip(*sorted(epas, reverse=True)))[0]
    divisions[division] = epas
```

```python
plt.figure(figsize=(20, 15))
for division, epas in dict(sorted(divisions.items(), key=itemgetter(0))).items():
    plt.plot(epas, '.-', label=division)
plt.legend()
plt.xlabel("Rank")
plt.ylabel("EPA")
plt.show()
```
